### ii. Fit GARCH(1, 1)

First we read in the data.

```python
stock = pd.read_parquet('../data/stock.parquet')
stock.head()
```

Now we calculate log returns.

```python
stock['log_ret'] = np.log(stock['adj_close']).diff()
stock.head()
```

The GARCH(1, 1) process is modeled as $$\sigma^2_t = \alpha_0 + \alpha_1 u^2_{t-1}+ \beta_1 \sigma^2_{t-1}.$$ My GARCH implementation estimates these parameters using Maximum Likelihood Estimation.

```python
res = get_garch_parameters(stock['log_ret'][1:])
```

The parameter estimates are shown below.

```python
alpha_0, alpha_1, beta_1 = res
print("Alpha_0 estimate:", alpha_0)
print("Alpha_1 estimate:", alpha_1)
print("Beta_1 estimate:", beta_1)
```
